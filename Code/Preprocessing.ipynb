```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def get_time_data(df, time, if_Month = False, if_Day = False, if_Hour = False, if_Minute = True):
    Year = time[0]
    Month = time[1]
    Day = time[2]
    Hour = time[3]
    Minute = time[4]
    df_year = df[df['Yr'] == Year]

    df_Month = df_year[df_year['M'] == Month]

    df_Day = df_Month[df_Month['D'] == Day]

    df_Hour = df_Day[df_Day['HH'] == Hour]

    df_Minute = df_Hour[df_Hour['MM'] == Minute]

    if if_Month:
        return df_Month
    if if_Day:
        return df_Day
    if if_Hour:
        return df_Hour
    if if_Minute:
        return df_Minute
```

```python
df = pd.read_csv(r'../Dataset/Automated_Traffic_Volume_Counts.csv')
```

```python
Year = 2019
Month = 8
Day = 14
Hour = 0
Minute = 0
Time = [Year, Month, Day, Hour, Minute]
df_time = get_time_data(df, Time, if_Month = True)
```
