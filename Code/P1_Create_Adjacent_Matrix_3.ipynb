```python
import pandas as pd
import numpy as np
import math
import os
```

```python
df_new = pd.read_csv('../Dataset/New_Road_Segments_Divided_Dir.csv')
```

```python
NF = df_new['NodeIDFrom']
NT = df_new['NodeIDTo']
ID = df_new['ID']

adj_f = []
adj_t = []
for i in range(180000, 228465):
    print(i)
    To = NT[i]
    df_To = df_new[df_new['NodeIDFrom'] == To]
    
    to = df_To['ID'].to_list()
    f = [ID[i]] * df_To.shape[0]
    
    adj_f = adj_f + f
    adj_t = adj_t + to
```

```python
adj_df = pd.DataFrame()

adj_df['From'] = adj_f
adj_df['To'] = adj_t

adj_df.to_csv('../Dataset/adjmatrix4.csv')
```
