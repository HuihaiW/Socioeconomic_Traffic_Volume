```python
def get_time_frame(df, M, D):
    df_M = df[df['M'] == M]
    df_D = df_M[df_M['D'] == D]
#     df_HH = df_D[df_D['HH'] == HH]
#     df_MM = df_HH[df_HH['MM'] == MM]
    return df_D
```
