```python
import pandas as pd
import numpy as np
import os
```

```python
df_worker = pd.read_csv(r'../Dataset/TrackInfo/Workers.csv')
df_occupation = pd.read_csv(r'../Dataset/TrackInfo/occupations.csv')
```

```python
df_worker = df_worker.drop(0)
```

```python
df_occupation = df_occupation.drop(0)
```

```python
worker_geo_id = df_worker['GEO_ID'].to_list()
occupation_geo_id = df_occupation['GEO_ID'].to_list()
```

```python
geo_id = []
for i in range(len(worker_geo_id)):
    geo_id.append(worker_geo_id[i].split('US')[1])
df_worker['S_GeoID'] = geo_id
df_worker = df_worker.astype({'S_GeoID': str})
df_worker.to_csv(r'../Dataset/TrackInfo/Workers_ID.csv')
```

```python
geo_id = []
for i in range(len(occupation_geo_id)):
    geo_id.append(occupation_geo_id[i].split('US')[1])
df_occupation['S_GeoID'] = geo_id
df_occupation = df_occupation.astype({'S_GeoID': str})
df_occupation.to_csv(r'../Dataset/TrackInfo/occupations_ID.csv')
```
