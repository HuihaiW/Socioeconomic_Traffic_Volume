```python
import pandas as pd
```

```python
df = pd.read_csv(r'../Data/Y.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.describe()
```

```python
a = df['0']
a = a[a!= -1].values
```

```python
a.mean()
a.max()
a.min()
a.std()
```

```python
mean = []
std = []
maximun = []
minimum = []
for n in df.columns:
    a = df[n]
    a = a[a!=-1].values
    mean.append(a.mean())
    maximun.append(a.max())
    minimum.append(a.min())
    std.append(a.std())
```
